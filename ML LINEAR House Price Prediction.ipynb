# Gradient Descent

```python
#STEP 1:initialise parameters
def param_initi(Y):
    
    m=0.1
    c=Y.mean()
    return m,c
```

```python
#STEP 2: generate predictions
def generate_predictions(m,c,X):
    predictions=[]
    for x in X:
        pred=(m*x)+c
        predictions.append(pred)
    return predictions
```

```python
#STEP 3:calculating cost
def compute_cost(predictions,Y):
    cost=np.sum(((predictions-Y)**2)/len(Y))
    return cost
```

```python
#STEP 4:updating parameters
def gradients(predictions,Y,X):
    n=len(Y)
    Gm=2/n*np.sum((predictions-Y)*X)
    Gc=2/n*np.sum((predictions-Y))
    return Gm,Gc
```

```python
def update_para(m_old,c_old,Gm_old,Gc_old,alpha):
    m_new=m_old-alpha*Gm_old
    c_new=c_old-alpha*Gc_old
    return m_new,c_new
```

```python
#final result pf gradient descent
def result(m,c,X,Y,cost,predictions,i):
#If the gradient desent converged to the optimum value before max_iter
    if i<max_iter -1:
        print("Gradient descent has converged at iteration {}".format(i))
    else:
        print("result after",max_iter,"iteration is:")
#plotting the final result
    plt.figure(figsize=(14,7),dpi=100)
    plt.scatter(X,Y,color='red',label='data points')
    label='final regression line: m={},c={}'.format(str(m), str(c))
    plt.plot(X,predictions,color='green',label=label)
    plt.xlabel("FLAT AREA")
    plt.ylabel("SALE PRICE")
    plt.title("FINAL REGRESSION LINE")
    plt.legend()
```

```python
#defining and reshaping the dataset
sale_price=small_data['sale_price'].values.reshape(-1,1)
flat_area=small_data['flat_area'].values.reshape(-1,1)
#declaring parameters
max_iter=1000
cost_old=0
alpha=0.01

#step1: intializing the parameters
m,c=param_initi(sale_price)

#Gradient descent in action
for i in range(0,max_iter):
#step 2:generating predictions
    predictions=generate_predictions(m,c,flat_area)
#step 3:calculating cost
    cost_new=compute_cost(predictions,sale_price)
#checking if GD is converged    
    if abs(cost_new-cost_old) <10**(-7):
        break
#calculating gradients    
    Gm,Gc=gradients(predictions,sale_price,flat_area)
#step 4:updating parameters m and c    
    m,c=update_para(m,c,Gm,Gc,alpha)
    #disply the results after every 20 iterations
    if i%20==0:
        print('After iteration',i,':m=',m,';c=',c,';cost_new',cost_new)
#updating cost_old
    cost_old=cost_new
    
#final results
result(m,c,flat_area,sale_price,cost_new,predictions,i)
```

```python
def MC_remover(data):
    vif=pd.Series([variance_inflation_factor(data.values,i) for i in range(data.shape[1])],index=data.columns)
    if vif.max()>5:
        print(vif[vif==vif.max()].index[0],'has been removed')
        data=data.drop(columns=(vif[vif==vif.max()].index[0]))
        return data
    else:
        print("NO multicollinearity id present anymore")
        return data
    
```

# Forward Selection

we are using forward selection right after the linear reg model

```python
def adj_r2(data_x,data_y,predictions):
    from sklearn.metrics import r2_score
    R=r2_score(data_y,predictions)
    n=len(data_y)
    m=len(data_x.columns)
    adj_r2=1-((n-1)*(1-R))/(n-m-1)
    return adj_r2
    
```

Here adj r2 < r2 marginally smaller,from this we can say that there are some features which are not adding value to final model

```python
#returns adj r2 corresponding to the model
def return_r2(X,Y,model):
    model.fit(X,Y)
    pred=model.predict(X)
    r2=adj_r2(X,Y,pred)
    return r2
```

```python
def forward_selection(model,X,Y,max_features):
    '''
    max_features:select best features up to max features value
    returns:
    RMAX:adj r2 of final model
    history:list of RMAX at every new ind var added to the model
    x[f_col]: best features upto max_features
    '''
    f_col=[] #stores the var being selected at every step of forwaed selection
    history=[]
    for i in range(max_features):
        f_col.append('blank')
        tmp=f_col[:]
        Rmax=-1 # tracks adding wich var produces max adj r2
        for var in X.columns:
            tmp[-1]=var #for any given iteration last element of tmp is assinged as a new column to track
            r2=return_r2(X[tmp],Y,model)
            if r2>=Rmax:
                Rmax=r2
                f_col[-1]=var
                elim=var
        print('selected',f_col[-1],'with new r2=',Rmax)
        history.append(Rmax)
    return Rmax,history,X[f_col]
        
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(normalize=True)
score,history,sel_data=forward_selection(lr,x_train,y_train,5)
```

# Backward Elimination

```python
def backward_elimination(model,X,Y,min_features):#specify no of features need to keep
    elim_data=X[:] #copy all the columns of x_train 
    r2_history=[] #to keep track adj r2 score hwen ever a feature is eliminated
    base_r2=return_r2(elim_data,Y,model)
    print('base_r2 is:',base_r2)
    for i in range(X.shape[1] - min_features):
        difference = -100
        
        for var in elim_data.columns:
            tmp = elim_data.drop(columns = [var])
            r2 = return_r2(tmp, Y, model)
            
            if (r2 - base_r2) > difference:
                difference = r2 - base_r2
                eliminate = var
                new_r2 = r2
        
        elim_data.drop(columns = [eliminate], inplace = True)
        r2_history.append(new_r2)
        
        print(eliminate,"Eliminated : New adj_r2 = ",new_r2,)
        base_r2 = new_r2
        
        return base_r2, r2_history, elim_data
```
